```python
# find distance

import numpy as np
x=np.random.randint(0,20,size=(100,1))
y=np.random.randint(0,10,size=(100,1))
# A=np.concatenate((x,y),axis=1)
A=np.block([x,y])
# print(A)
B=A[0]
C=B-A

D=C**2

x=np.round(np.sqrt(D[:,0]+D[:,1]),3)
# print(x.reshape(100,1))
```

```python
##create a ring in the plot 

import numpy as np
import matplotlib.pyplot as plt

#random data
x =np.random.rand(1000)
y = np.random.rand(1000)

#reshaping for blocking
x=x.reshape(1000,1) 
y=y.reshape(1000,1)
#making pair
A=np.block([x,y])
#plotting points
plt.scatter(x, y,color="blue",s=10)

#radius
r_in=0.25
r_out=0.3
#mid point
B=A[999]

#finding distance
dist=np.sqrt((x-B[0])**2+(y-B[1])**2)

#finding and plotting the ring
points=np.logical_and(dist<r_out,dist>r_in)
plt.scatter(x[points],y[points],color='red',s=10)

#coloring inside points
points_inside=dist<r_in
plt.scatter(x[points_inside],y[points_inside],color='black',s=10)


plt.scatter(B[0],B[1],color='yellow',s=5)
```
