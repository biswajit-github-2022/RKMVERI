```python
import numpy as np
x=np.random.randint(0,20,size=(100,1))
y=np.random.randint(0,10,size=(100,1))
# A=np.concatenate((x,y),axis=1)
A=np.block([x,y])
# print(A)
B=A[0]
C=B-A

D=C**2

x=np.round(np.sqrt(D[:,0]+D[:,1]),3)
print(x.reshape(100,1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

#random data
x =np.random.rand(1000)
y = np.random.rand(1000)

#reshaping for blocking
x=x.reshape(1000,1) 
y=y.reshape(1000,1)
#making pair
A=np.block([x,y])
#plotting points
plt.scatter(x, y,color="blue",s=10)

#radius
r_in=0.25
r_out=0.3
#mid point
B=[0.5,0.5]

#finding distance
C=B-A
D=C**2
z=np.round(np.sqrt(D[:,0]+D[:,1]),5)
# print(z)
# print(z.reshape(1000,1))

#creating circle
for i in range(len(z)):
    if r_in<z[i]<r_out:
        plt.scatter(x[i],y[i],color='red',s=15)
```
