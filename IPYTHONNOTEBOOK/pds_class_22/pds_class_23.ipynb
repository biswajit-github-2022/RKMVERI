```python
import numpy as np
import pandas as pd
```

```python
x=np.random.randint(1,10,(4,6))
print(x)
col_no=2
t=x[:,col_no-1:col_no]
print(t)
for i in range(x.shape[1]+1):
    if col_no==i:
        continue
    x[:,i-1:i]=x[:,i-1:i]-t
print(x)
```

```python
x=np.random.randint(1,10,(4,6))
print(x)
row_no=2
t=x[row_no-1,:]
print(t)
for i in range(x.shape[0]+1):
    if row_no==i:
        continue
    x[i-1,:]=x[i-1,:]-t
print(x)
```

```python
x=np.random.randint(1,10,(6,6))
print(x)
a=np.eye(6,6)
print(x[a==1])
x[:,0:(x.shape[0]-1)]=x[:,(x.shape[0]-1):0]
print(x[:,0:(x.shape[0]-1)])
print(x)
print(x[a==1])
```
