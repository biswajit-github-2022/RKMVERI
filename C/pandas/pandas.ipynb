```python
def join_csv(data1,data2):
    result = pd.merge(data1, data2)
    return result
join_csv(data1,data2)
```

```python
data3=join_csv(data1,data2)
def select_op(roll:int):
    l=[]
    for i in range(data3.shape[0]):
        if data3.iloc[i]['roll']==roll:
            l.append(data3.iloc[i])
    return l
def select_op_name(name:str):
    l=[]
    for i in range(data3.shape[0]):
        if data3.iloc[i]['name']==name:
            l.append(data3.iloc[i])
    return l
def select_op_marks(marks:int):
    l=[]
    for i in range(data3.shape[0]):
        if data3.iloc[i]['marks']==marks:
            l.append(data3.iloc[i])
    return l
x=select_op_marks(67)
df=pd.DataFrame(x)
# print(df)
# print(data3.iloc[1])
df.to_csv("csv3.csv",index=False)
```

# Threading

```python
import threading
import time

# Define a simple function that will run in a thread
def print_numbers():
    for i in range(5):
        time.sleep(1)  # Simulate some work
        print(f"Thread 1: {i}")

def print_letters():
    for letter in 'ABCDE':
        time.sleep(1)
        print(f"Thread 2: {letter}")

# Create thread objects
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)

# Start the threads
thread2.start()
thread1.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

print("Both threads have finished.")
```
